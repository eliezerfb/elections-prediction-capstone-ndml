```python
import pandas as pd
import platform
from IPython.display import display, HTML
```

```python
pd.options.display.float_format = '{:.2f}'.format
pd.set_option('display.max_columns', None)
```

```python
if platform.os == 'Windows':
    path = 'Data\\'
else:
    path = 'Data//'
```

```python
df = pd.read_csv(path+'dados_tratados.csv', encoding='latin1')
```

```python
s = {'SUPLENTE':0, 'ELEITO POR QP':1, 'NÃO ELEITO':0, 'ELEITO POR MÉDIA':1}
df['DESC_SIT_TOT_TURNO'] = df['DESC_SIT_TOT_TURNO'].map(s)
```

```python
df_federal_ro = df[(df['SIGLA_UF'] == 'RO') & (df['CODIGO_CARGO'] == 6)].copy() 
print(df_federal_ro[df_federal_ro['DESC_SIT_TOT_TURNO'] == 1]['DESC_SIT_TOT_TURNO'].shape)
nome_pesq_eleitos = ['MARINHA',
                     'NILTON', 
                     'MARCOS ROGERIO', 
                     'AGNALDO MUNIZ', 
                     'MOSQUINI',
                     'BIANCO',
                     'LINDOMAR',
                     'RAQUEL DONADON']
df_federal_ro['PESQUISA'] = 0

for nome_pesq_eleito in nome_pesq_eleitos:
    df_federal_ro.loc[df_federal_ro['NOME_CANDIDATO'].str.contains(nome_pesq_eleito), 'PESQUISA'] = 1
    
display(df_federal_ro[df_federal_ro['PESQUISA'] == 1])
```

```python
from sklearn.metrics import *
def metrics(y_test, y_pred):
    print('acc:', accuracy_score(y_test, y_pred))
    print('precision: ', precision_score(y_test, y_pred))
    print('recall: ', recall_score(y_test, y_pred))
    print('f1:', f1_score(y_test, y_pred))
    print('ROC:', roc_auc_score(y_test, y_pred))
```

```python
metrics(df_federal_ro['PESQUISA'], df_federal_ro['DESC_SIT_TOT_TURNO'])
```
